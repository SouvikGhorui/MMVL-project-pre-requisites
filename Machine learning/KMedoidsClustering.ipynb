```python
import kmedoids
import numpy as np
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
X = np.random.rand(50, 2)
X
```

```python
dist_matrix = pairwise_distances(X, metric = 'euclidean')
dist_matrix
```

```python
km = kmedoids.KMedoids(n_clusters= 3)
km.fit(dist_matrix)
labels = km.labels_
medoids_indices = km.medoid_indices_
inertia = km.inertia_
```

```python
print(f'Cluster labels: {labels}')
print(f'Medoid indices: {medoids_indices}')
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X[:,0], X[:,1], marker = 'o', label = 'Data points', s = 50)

medoids = X[medoids_indices]
plt.scatter(medoids[:,0], medoids[:,1], color = 'red', marker = 'X', label = 'Medoid', s = 200)


```
