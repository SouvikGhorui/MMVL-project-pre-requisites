```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

We use circle dataset to test why we use `Density Based Clustering` instead of `K means clustering`

```python
from sklearn.datasets import make_circles
X, _ = make_circles(n_samples = 200)
```

Plot the dataset 

```python
plt.scatter(X[:,0], X[:,1], cmap='viridis', marker='o')
```

First we test the dataset with `K means cluster`

```python

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
y_predicted = kmeans.fit_predict(X)

```

plot after the clustering 

```python
plt.scatter(X[:,0], X[:,1], c = y_predicted )
```

From the figure, it is clear that the `K means cluster` is not good for this kind of dataset. So, we use `Density based clustering`. 

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps = 0.1, min_samples=3)
y_predicted = db.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c = y_predicted)
```

Here we can see the `DBSCAN` method is more powerfull than `K means clustering`
