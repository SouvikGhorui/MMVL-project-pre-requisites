```python
from skimage import data, exposure, io
from skimage.feature import hog
import matplotlib.pyplot as plt 
```

Load image to use HOG


```python
image = io.imread(r"C:\Users\sghor\Desktop\Souvik\Programming\MMVL project requirments\Image processing\Photos\test_flower.jpg", as_gray = True)
```

```python
fd, hog_image = hog(image, orientations= 8 ,pixels_per_cell= (16, 16), cells_per_block=(1,1), visualize= True)
hog_scaled = exposure.rescale_intensity(hog_image, in_range= (0,10))
```

```python
plt.figure(figsize = (20, 10))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(hog_scaled, cmap='gray')
```
